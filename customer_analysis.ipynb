```python
#load file
import pandas as pd
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
#check starting rows
df.head()
```

```python
#stats of all columns
df.describe(include='all')
```

```python
#check missiing values
df.isnull().sum()
```

```python
#replace misisng review rating with missinbg val of all items of that category
df['review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
#recheck missiing values(data cleaning)
df.isnull().sum()
```

```python
#lowercasing for even naming
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#creating new column of age_group (split age into 4 equal size grps)
labels = ['Young_adult','Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
#looking age groups
df[['age','age_group']].head(10)
```

```python
#purchase frequency days feature(transform textual freq to nos)
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
df['purchase_frequency_days']= df['frequency_of_purchases'].map(frequency_mapping)
```

```python
#checking transformed frequency
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
# Dropping promo code used column

df = df.drop('promo_code_used', axis=1)
```
